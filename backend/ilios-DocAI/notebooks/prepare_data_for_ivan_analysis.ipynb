```python
from src.pipelines.constants import NOT_PROVIDED_STR

data = data.copy().dropna(
        subset=["Legal Terms", "Predicted Legal Terms"]
    )
y_true = data["Legal Terms"] != NOT_PROVIDED_STR
y_pred = data["Predicted Legal Terms"] != NOT_PROVIDED_STR
true_positive = y_true & y_pred

data = (data[true_positive][[ 
    "file_name", "key_item",
    "Legal Terms", "Predicted Legal Terms",
    
    "bleu_score", "levenshtein_score", "rouge1_f1_score"]].sort_values("bleu_score", ascending=False))
 
 
# data = data[data["bleu_score"] > 0.0]
# data.to_csv("data_for_analysis.csv", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the normal distribution
mean = 110
std_dev = 50  # You can adjust this value to tune the width of the parabola

# Generate a sequence of numbers from 0 to 250
x = np.linspace(0, len(data), len(data))

# Calculate the normal distribution
y = (1 / (np.sqrt(2 * np.pi * std_dev**2))) * np.exp(-((x - mean)**2) / (2 * std_dev**2))

# Plot the normal distribution

fig, ax1 = plt.subplots()

ax1.plot(x, y)

ax2 = ax1.twinx()

ax2.plot(data["bleu_score"].values, color='red')

plt.title('Normal Distribution')
# ax1.ylabel('Probability Density')
plt.show()
```
