This notebook contains _llm_evaluation_ function testing to make it more robust for different model outputs.

```python
import pytest
from src.validation.validation import llm_evaluation, get_evaluation_llm_chain

@pytest.mark.parametrize(
    "reference_answer, new_answer, expected",
    [
        ("Not provided.", "Not provided.", 2.0),
        ("Not provided.", "Not provided", 2.0),
        ("Not provided.", 'The document does not appear to contain the term "Mechanical Completion Date" or any direct references to that specific term. There are no relevant quotes or sections that mention mechanical completion.', 2.0),
        ("Not provided.", 'The term "Degradation Amount" is not explicitly mentioned in the given document. There are no direct citations related to this term.', 2.0),
        ("Not provided.", 'The document does not appear to contain the term "Site Lease Prepayment (Y/N)". There are no direct citations relevant to this term. So the answer is: \n Not provided.', 2.0),
    ],
)
def test_llm_evaluation(reference_answer, new_answer, expected):
    llm_chain = get_evaluation_llm_chain()
    result = llm_evaluation(reference_answer, new_answer, llm_chain)
    assert result == expected
```
