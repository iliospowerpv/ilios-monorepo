````python
import pandas as pd
````

````python
from src.pipeline.utils import get_terms_and_definitions

terms_and_definitions = get_terms_and_definitions()
````

````python
for _, term_definition_row in terms_and_definitions.iterrows():
    term = term_definition_row["Key Items"]
    definition = term_definition_row["Definitions"]
    examples = [
        term_definition_row[k]
        for k in term_definition_row.keys()
        if ("example" in k) and not pd.isna(term_definition_row[k])
    ]
    
    print("Act as a senior prompt engineer. Having this term, definition and examples of text to look for, you can start to look for the term in the text and understand the context.")
    print(f"Term: {term}")
    print(f"Definition: {definition}")
    print(f"Examples of text to look for:")
    for example in examples:
        print(f"```\n- {example}")
    
    
    print("-"*50)
````
