```python
import pandas as pd
```

```python
post_process_df = pd.read_csv("../2024-04-15_09-05-34_output-site-lease-with-preprocessing.csv")
no_post_process_df = pd.read_csv("../2024-04-15_12-53-27_output-site-lease-no-postprocess.csv")
# 2024-04-15_13-19-22_output-site-lease-post-process-95.csv
```

```python
post_process_df.dropna(subset=["Predicted Legal Terms", "Legal Terms"], inplace=True)
no_post_process_df.dropna(subset=["Predicted Legal Terms", "Legal Terms"], inplace=True)
```

```python

from src.pipelines.constants import NOT_PROVIDED_STR


def get_true_positive(results_df):
    y_true = results_df["Legal Terms"] != NOT_PROVIDED_STR
    y_pred = results_df["Predicted Legal Terms"] != NOT_PROVIDED_STR
    return y_true & y_pred
    

post_process_df = post_process_df[get_true_positive(post_process_df)]
no_post_process_df = no_post_process_df[get_true_positive(no_post_process_df)]


post_process_df.shape, no_post_process_df.shape
```

```python
cols = ['file_name', 'key_item', 'Legal Terms', 'Predicted Legal Terms', "bleu_score"]

no_post_process_df[cols].merge(
    post_process_df[cols],
    on=['file_name', 'key_item'],
    suffixes=("_no_post_process", "_post_process"),
    how="outer").sort_values("bleu_score_post_process", ascending=False)
```
